````python
# Let's first start with importing the necessities

import polars as pl
import datetime as dt
````

````python
df = pl.DataFrame(
    {
        "name" : ['Ben','Jerry','Bob','John','Alice'],
        "birthday" : [
            dt.date(1999,4,12),
            dt.date(1996,5,11),
            dt.date(2000,7,15),
            dt.date(2001,10,22),
            dt.date(2005,10,9)
                      ],

        "age" : [32,15,26,33,25],
        "height" : [1.67,1.76,1.66,1.87,1.55]
    }
)

print(df)
````

````python
# Expressions are one of the main strengths of Polars because they provide a modular and flexible way of expressing data transformer.

# Here is a quick example -

pl.col('age')/ (pl.col('height') * 10)
````

````python
# Now let's handle some data manipulation

result_df = df.select(
    pl.col('name'),
    pl.col('birthday').dt.year().alias('Birth_Year'),
    (pl.col('age')/ (pl.col('height') * 10)).alias("Useless_Metric")
)

print(result_df)
````

````python
exp_df = pl.DataFrame(
    {
    "Name" : ['Rutuj','Shreya','Kalki','Sarfaraz'],
    "Age" : [24,21,19,29],
    "Sex" : ['M','F','M','M'],
    "Height": [177,163,186,180],
    "Weight": [145,170,200,230],
    "Major": ['Computer Science','DataBase Designer','World Destroyer','Bio-Medical'],
    "Hobbies": ['Gaming', 'Talking About Love and Marriage', 'Fixing the Spiritual Realm', 'Hitting on guys']
}
)

print(exp_df)
````

````python
df_mod = exp_df.select(
    pl.col('Name'),
    pl.col('Hobbies'),
    (pl.col('Height')/ pl.col('Weight')).alias('Height by Weight Ration'),
    pl.col('Age').alias('This years old')
)

print(df_mod)
````

````python
df_mod_2 = exp_df.select(
    pl.col('Name'),
    (pl.col('Weight','Height') * 0.90).round(2).name.suffix(" -10%")
)
# This is so simple and easy to handle
print(df_mod_2)
````

````python
# Now lets say with modifying Columns, you want to add some fresh new columns, how would we do that?

print("This is how: ")

df_mod_3 = exp_df.with_columns(
    Weight_in_Kilos = (pl.col('Weight') * 0.45),
    Bmi = (pl.col('Weight')* 4000) / (pl.col('Height') ** 2)
)
# This is not the realistic formula for generating the BMI, but its now some-what shows realistic data.
print(df_mod_3)
````

````python
res = df.filter(pl.col('birthday').dt.year() < 2000)
print(res)
# This is how we use the filter operator
````

````python
# Let's try something like this again

res = df.filter(
    pl.col('birthday').is_between(dt.date(1982,12,31), dt.date(2106,1,1)),
    pl.col('height') > 1.70
)

print(res)
````

````python
# Let us now learn about the group by function, for that lets us first create a relevant dataframe



df_sp = pl.DataFrame(
    {
        "Years": [1999,1998,1997,2000,1999,1998,1999,2001,2000],
        "Revenue": [4000,2500,5000,2400,1545,4999,7000,5000,5000]
    }
)

res = (
    df_sp
    .group_by("Years", maintain_order=True)
    .agg(
        pl.col("Revenue").mean().round(3).alias("Total Profit")
    )
)

print(res)

# for us to see the names of the columns here it is:
print(repr(df_sp.columns))
````

````python
# Let's create another dataframe

df = pl.DataFrame(
    {
        "name": ["Alice Archer", "Ben Brown", "Chloe Cooper", "Daniel Donovan","Raphael Soliciting", "Cancer Retrospect","Unicorn Marine"],
        "birthdate": [
            dt.date(1997, 1, 10),
            dt.date(1935, 2, 15),
            dt.date(1983, 3, 22),
            dt.date(1981, 4, 30),
            dt.date(2006, 6, 14),
            dt.date(1968, 5, 15),
            dt.date(2000, 8, 12)
        ],
        "weight": [57.9, 72.5, 53.6, 83.1, 77.3, 55.9, 99.9],  # (kg)
        "height": [1.56, 1.77, 1.65, 1.75, 1.77, 1.54, 1.89],  # (m)
    }
)

print(df)
````

````python
# Let's do quite a complex query

result = df.with_columns(
    (pl.col("birthdate").dt.year() // 10 * 10).alias("Decade"), # Converting the BirthDate column to a new column namely Decade while making changes to it
    pl.col('name').str.split(by=" ").list.first(), # Splitting the full names so that only the first name is shown in the output
).select(
    pl.all().exclude("birthdate"),
).group_by(
    pl.col("Decade"),
    maintain_order = True
).agg(
    pl.col("name"),
    pl.col("weight","height").mean().round(2).name.prefix("avg_")
)

print(result)
````

````python
# Let us try combining two DataFrames

df2 = pl.DataFrame({
        "name": ["Ben Brown", "Daniel Donovan", "Alice Archer", "Harry Styles","Chloe Cooper"],
        "parent": [True, False, False, False, True],
        "siblings": [1, 2, 3, 4,5],

})

print(df.join(df2, on = "name", how = "left"))
````

````python
# Let's try to do a right join

print(df.join(df2, on= "name", how= "right"))
````

````python
# We Could Also Concatenate DataFrames

df3 = pl.DataFrame({
        "name": ["Ethan Edwards", "Fiona Foster", "Grace Gibson", "Henry Harris"],
        "birthdate": [
            dt.date(1977, 5, 10),
            dt.date(1975, 6, 23),
            dt.date(1973, 7, 22),
            dt.date(1971, 8, 3),
        ],
        "weight": [67.9, 72.5, 57.6, 93.1],  # (kg)
        "height": [1.76, 1.6, 1.66, 1.8],


})

print(pl.concat([df,df3], how="vertical"))
````

# ğŸ§© Polars â€” Additional Features & Extensions

This section lists advanced **Polars feature flags** and **optional modules**

to high-performance optimizations and specialized operations âš¡

---

## ğŸ§± Additional Data Types

- **dtype-date** â€” Support for calendar dates
- **dtype-datetime** â€” Support for full timestamp data
- **dtype-time** â€” Time-only data representation
- **dtype-duration** â€” Represent durations or timedeltas
- **dtype-i8 / dtype-i16** â€” 8-bit and 16-bit signed integers
- **dtype-u8 / dtype-u16** â€” 8-bit and 16-bit unsigned integers
- **dtype-categorical** â€” Efficient categorical encoding
- **dtype-struct** â€” Support for nested struct fields

---

## ğŸ§  Lazy Computation & Query Optimization

- **lazy** â€” Lazy API for deferred execution and query planning
- **regex** â€” Regular expressions in column selection
- **dot_diagram** â€” Visualize lazy query plans as DOT diagrams
- **sql** â€” Run SQL queries directly on Polars DataFrames
- **streaming** â€” Process datasets larger than available RAM

---

## ğŸ² Randomness & Array Interoperability

- **random** â€” Generate arrays with randomly sampled values
- **ndarray** â€” Convert from DataFrame to NumPy ndarray
- **temporal** â€” Seamless conversion between Chrono and Polars temporal types
- **timezones** â€” Enable timezone support
- **strings** â€” Extra string manipulation utilities
  - **string_pad** â€” Padding operations (`pad_start`, `pad_end`, `zfill`)
  - **string_to_integer** â€” Parse integer values from strings
- **object** â€” Support for generic `ObjectChunked<T>` arrays

---

## âš™ï¸ Performance-Related Features

- **nightly** â€” Nightly-only optimizations like SIMD and specialization
- **performant** â€” Enables faster paths (slightly longer compile times)
- **bigidx** â€” Use 64-bit indices for very large datasets (`>> 2^32` rows)
- **cse** â€” Common Subplan Elimination for query optimization

---

## ğŸ“¦ IO & Serialization Features

- **serde / serde-lazy** â€” Serialize & deserialize using Serde (JSON and more)
- **parquet** â€” Read and write Apache Parquet format
- **json** â€” JSON serialization support
- **ipc** â€” Support for Arrow IPC (Feather) format
- **decompress** â€” Automatically decompress CSVs (supports: gzip, zlib, zstd)

---

## ğŸ§® DataFrame-Level Operations
