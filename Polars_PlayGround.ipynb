````python
# Let's first start with importing the necessities

import polars as pl
import datetime as dt
````

````python
df = pl.DataFrame(
    {
        "name" : ['Ben','Jerry','Bob','John','Alice'],
        "birthday" : [
            dt.date(1999,4,12),
            dt.date(1996,5,11),
            dt.date(2000,7,15),
            dt.date(2001,10,22),
            dt.date(2005,10,9)
                      ],

        "age" : [32,15,26,33,25],
        "height" : [1.67,1.76,1.66,1.87,1.55]
    }
)

print(df)
````

````python
df.write_csv("/Users/swapnanilbala/Desktop/polars_template.csv") # writing the dataframe into this particular file and then calling it back again
df_csv = pl.read_csv("/Users/swapnanilbala/Desktop/polars_template.csv", try_parse_dates= True)
print(df_csv)
````

````python
# Expressions are one of the main strengths of Polars because they provide a modular and flexible way of expressing data transformer.

# Here is a quick example -

pl.col('age')/ (pl.col('height') * 10)
````

````python
# Now let's handle some data manipulation

result_df = df.select(
    pl.col('name'),
    pl.col('birthday').dt.year().alias('Birth_Year'),
    (pl.col('age')/ (pl.col('height') * 10)).alias("Useless_Metric")
)

print(result_df)
````

````python
exp_df = pl.DataFrame(
    {
    "Name" : ['Rutuj','Shreya','Kalki','Sarfaraz'],
    "Age" : [24,21,19,29],
    "Sex" : ['M','F','M','M'],
    "Height": [177,163,186,180],
    "Weight": [145,170,200,230],
    "Major": ['Computer Science','DataBase Designer','World Destroyer','Bio-Medical'],
    "Hobbies": ['Gaming', 'Talking About Love and Marriage', 'Fixing the Spiritual Realm', 'Hitting on guys']
}
)

print(exp_df)
````

````python
df_mod = exp_df.select(
    pl.col('Name'),
    pl.col('Hobbies'),
    (pl.col('Height')/ pl.col('Weight')).alias('Height by Weight Ration'),
    pl.col('Age').alias('This years old')
)

print(df_mod)
````

````python
df_mod_2 = exp_df.select(
    pl.col('Name'),
    (pl.col('Weight','Height') * 0.90).round(2).name.suffix(" -10%")
)
# This is so simple and easy to handle
print(df_mod_2)
````

````python
# Now lets say with modifying Columns, you want to add some fresh new columns, how would we do that?

print("This is how: ")

df_mod_3 = exp_df.with_columns(
    Weight_in_Kilos = (pl.col('Weight') * 0.45),
    Bmi = (pl.col('Weight')* 4000) / (pl.col('Height') ** 2)
)
# This is not the realistic formula for generating the BMI, but its now some-what shows realistic data.
print(df_mod_3)
````

````python
res = df.filter(pl.col('birthday').dt.year() < 2000)
print(res)
# This is how we use the filter operator
````
